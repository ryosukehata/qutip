```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation
from mpl_toolkits.mplot3d import Axes3D
from qutip import Bloch
from qutip.ipynbtools import plot_animation
```

```python
data_dir = "../03_Bloch"
```

```python
b = Bloch()
b.show()
b.save(data_dir+"/empty_Bloch.png")
```

```python
from qutip import basis
from scipy import sqrt
a = 0.5
```

# アダマールゲートを|0>に作用させる

```python
phi0 = (a* basis(2,0) + (1-a)*basis(2,1))/(sqrt(a**2 + (1-a)**2))
b = Bloch()
b.add_states([phi0.unit()])
b.show()
b.save(data_dir+"/+x_Bloch.png")
```

# アダマールゲートを|1>に作用させる

```python
a = 0.5
phi0 = (a* basis(2,0) - (1-a)*basis(2,1))/(sqrt(a**2 + (1-a)**2))
b = Bloch()
b.add_states([phi0.unit()])
b.show()
b.save(data_dir+"/-x_Bloch.png")
```

# 位相ゲートを|1>に作用させる

```python
import numpy as np
from numpy import pi, exp
```

```python
theta = pi/4
phase = exp(-1j*theta)
a=0
phi0 = (a* basis(2,0) + phase*(1-a)*basis(2,1))/(sqrt(a**2 + (1-a)**2))
b = Bloch()
b.add_states([phi0.unit()])
b.show()
b.save(data_dir+"/phase_Bloch1.png")
```

```python
# アダマールゲートの後
theta = pi/4
phase = exp(-1j*theta)
a=0.5
phi0 = (a* basis(2,0) + phase*(1-a)*basis(2,1))/(sqrt(a**2 + (1-a)**2))
b = Bloch()
b.add_states([phi0.unit()])
b.show()
b.save(data_dir+"/phase_Bloch2.png")
```
